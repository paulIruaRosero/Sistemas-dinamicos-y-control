```python
# Funciones matematicas
import numpy as n
# Funciones para graficos
import matplotlib.pyplot as plt
# Funciones parecidas a Matlab
from control.matlab import *
#definicion del sistema
num=[1,1]
den=[1,2,2]
ftla= tf(num, den)
print("Funcion de transferencia en LA de la planta:")
print(ftla)
ftlc=feedback(ftla,1)
print("Funcion de transferencia en lazo cerrado: ")
print(ftlc)
```

```python
# Grafico de polos y ceros del sistema en lazo cerrado.
plt.figure(1, figsize=[7, 4])
pzmap(ftlc)
plt.xlabel("Real")
plt.ylabel("maginario")
plt.title("Plano complejo. Polos del sistema en LC")
plt.show(block=False)
```

**SimulaciÃ³n del comportamiento en lazo abierto y cerrado para entrada impulso y escalÃ³n unitario**

```python
# Respuesta escalon del sistema en LA y LC
plt.figure(2, figsize=[6, 4])
yout, T = step(ftla, T=25)
yout1, T1 = step(ftlc, T=25)
plt.plot(T.T, yout.T,T1.T,yout1.T)
plt.xlabel("tiempo (s)")
plt.ylabel("$y(t)$")
plt.title("Respuesta escalÃ³n sistema en LA y LC")
plt.grid()
```

```python
# Respuesta impulso del sistema en LA y LC
plt.figure(3, figsize=[6, 4])
yout, T = impulse(ftla, T=25)
yout1, T1 = impulse(ftlc, T=25)
plt.plot(T.T, yout.T,T1.T,yout1.T)
plt.xlabel("tiempo (s)")
plt.ylabel("$y(t)$")
plt.title("Respuesta impulso sistema en LA y LC")
plt.grid()
plt.show()
```
