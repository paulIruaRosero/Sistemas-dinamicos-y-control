**diferentes tipos de excitaciones utilizando Python-control.matlab**

```python
import numpy as np
import matplotlib.pyplot as plt
import control.matlab as ctl

# Definimos una funciÃ³n de transferencia G(s) = 1 / (s^2 + 2s + 2)
num = [1]
den = [1, 2, 2]
G = ctl.tf(num, den)

# Definimos el tiempo de simulaciÃ³n
t = np.linspace(0, 10, 1000)

# --- Excitaciones ---
# EscalÃ³n
t1, y1 = ctl.step(G, T=t)

# Impulso
t2, y2 = ctl.impulse(G, T=t)

# Rampa (usamos lsim con u=t)
u_rampa = t
t3, y3, _ = ctl.lsim(G, U=u_rampa, T=t)

# Senoidal
u_seno = np.sin(2*t)
t4, y4, _ = ctl.lsim(G, U=u_seno, T=t)

# --- GrÃ¡ficas ---
plt.figure(figsize=(10,8))

plt.subplot(2,2,1)
plt.plot(t1, y1)
plt.title("Respuesta al EscalÃ³n")
plt.grid()

plt.subplot(2,2,2)
plt.plot(t2, y2)
plt.title("Respuesta al Impulso")
plt.grid()

plt.subplot(2,2,3)
plt.plot(t3, y3, label="Salida")
plt.plot(t3, u_rampa, 'r--', label="Entrada Rampa")
plt.title("Respuesta a la Rampa")
plt.legend()
plt.grid()

plt.subplot(2,2,4)
plt.plot(t4, y4, label="Salida")
plt.plot(t4, u_seno, 'r--', label="Entrada Senoidal")
plt.title("Respuesta a la Senoidal")
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```
